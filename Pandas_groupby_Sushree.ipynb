## load a new data file u.user and read it

```python
u_user_file=pd.read_csv("C:/Users/sushr/Downloads/u.user",sep="|")
u_user_file
```

### Discover what is the mean age per occupation

```python
u_user_file.groupby("occupation")["age"].mean()
```

### Discover the Male ratio per occupation and sort it from the most to the least

```python
# function for new column in which only male should be counted
def genderratio(x):
    if x == "M":
        return 1
    else:
        return 0


u_user_file["gender_type"] = u_user_file["gender"].apply(genderratio)
Male_ratio_per_occup = u_user_file.groupby(
    "occupation")["gender_type"].sum()/u_user_file["occupation"].value_counts()*100
Male_ratio_per_occup.sort_values(ascending=False)
```

### For each occupation, calculate the minimum and maximum ages

```python
u_user_file.groupby("occupation")["age"].agg(["min","max"])
```

### For each combination of occupation and gender, calculate the mean age without heirarchical indexing

```python
u_user_file.groupby(["occupation","gender"])["age"].mean()
```

```python
u_user_file.groupby(["occupation","gender"])["age"].mean().unstack()
```
